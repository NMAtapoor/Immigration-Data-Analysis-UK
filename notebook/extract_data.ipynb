# Extract data-sheets from excel file

```python
import pandas as pd
from pathlib import Path

DATA_PATH = Path("../data/rawData/entry-clearance-visa-outcomes-datasets-sep-2025.xlsx")
DATA_PATH2 = Path("../data/cleanData/entry-clearance-visa_application.csv")
data01 = pd.read_excel(DATA_PATH, 
              sheet_name="Data_Vis_D01", 
              skiprows = 3,
              header = 0,
              nrows=100)

data02 = pd.read_excel(DATA_PATH, 
              sheet_name="Data_Vis_D02", 
              skiprows = 3,
              header = 0,
              nrows=100)
```

```python
data01.to_csv(
    DATA_PATH2,
    index=False,
    encoding="utf-8-sig"
)
```
